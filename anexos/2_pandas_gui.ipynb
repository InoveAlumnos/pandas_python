```python
import pandas as pd

# Módulo que contiene el dataset de Iris
import sklearn
from sklearn.datasets import load_iris
sklearn.__version__
```

# Recolectar datos
<div align="center"><img src="https://raw.githubusercontent.com/InoveAlumnos/dataset_analytics_python/master/images/Pipeline1.png" width="1000" align="middle"></div>

### El dataset(Conjunto de datos) de Iris es un dataset ya cargado en la librería de sklearn, se utiliza para la práctica de preprocesamiento de datos. Solo basta con importarlo para comenzar a utlizarlo.

```python
# Se carga la información en la variable iris
iris = load_iris()

# Se construye el DataFrame llamado df_data, considerando los datos y las columnas que posee el dataset.
df_data = pd.DataFrame(data= iris.data, columns= iris.feature_names)
df_data 
```

```python
# Se construye un DataFrame con la información categorica. Es decir, una columna llamada species cuyos registros son números que representan una categoría,
# La especia de cada flor.
df_target = pd.DataFrame(data= iris.target, columns= ['species'])
df_target
```

```python
# Se concatena ambos DataFrame con el método concat() de Pandas
# Se indica como primer parámetro la lista de DataFrame y como segundo parámetro axis=1 para que
# la información se construya fila a fila.
df = pd.concat([df_data, df_target], axis= 1)
df.head()
```
